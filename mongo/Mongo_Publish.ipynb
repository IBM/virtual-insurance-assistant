```python
# Convert address to coords
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="test")
address = "200 N Spring St, Los Angeles, CA 90012"
user_location = geolocator.geocode(address)
user_coords = [user_location.longitude, user_location.latitude]
```

```python
# Generate a random location for each business (if using synthetic dataset)
random.seed(None)
long, lat = user_location.longitude, user_location.latitude
degrees = 1.0
gen_locs = lambda x: x + random.randrange(-40,100, 1) / 100
locs = []
for b in businesses: 
    new_loc = [gen_locs(lat), gen_locs(long)]
    b['location'] = { "type": "Point", "coordinates": new_loc }
```

```python
# Filter mechanic shops that are in a location with a specific radius
meters = 10000
nearby_mechs = db.businesses_collection.aggregate(
    [{
        "$geoNear": {
            "near": { "type": "Point", "coordinates": user_coords },            
            "distanceField": "calculated",
            "includeLocs": "location",
            "maxDistance": meters,
            "spherical": False            
        }
    }]
)
```

```python
# get ids of best performing mechanics selected in previous cell
nearby_mech_ids = list(map(lambda d: d['id'], list(nearby_mechs)))
nearby_mech_ids
```

```python
# perform aggregation to select top X mechanics
repair_type = "Maintenance"
top_mechs = db.reviews_collection.aggregate(
  [      
    {
       "$match": {
         "entities.type": 'Work_Type', 
         "entities.disambiguation.subtype": repair_type,
         "business_id": {"$in": nearby_mech_ids},
         "$or": [ {"sentiment": "positive"}, {"sentiment": "neutral"} ]
        }
    },
    {
      "$group" :
        {
          "_id" : "$business_id",
          "repairCount": { "$sum": 1 },
        }
    },
    { 
      "$lookup": {
        "from": "businesses_collection",
        "localField": "_id",
        "foreignField": "id",
        "as": "business"
      }
    },
    {
      "$replaceRoot": 
        { 
          "newRoot": 
            { "$mergeObjects": [ { "$arrayElemAt": [ "$business", 0 ] }, "$$ROOT" ] 
          } 
        }
    },  
    { 
      "$project": { 
          "_id": 1,
          "repairCount": 1, 
          "name": 1,
          "location": 1
      }
    },
    { 
      "$sort" : { 
          "repairCount" : -1
      }
    }      
   ]
 )
```
